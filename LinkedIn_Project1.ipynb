I've decided to perform this process later on in order to make sure that all the data that I would get would be from approximately the same time. It may differ from 1 to 3 minutes time depending on the amount of pages that had to be visited during this period, however it is still useful data to be able to compare.

First data to be stored will be the data included on the main page of the website. This data correspond to statistics from the whole world constantly being displayed there. It will be necessary, in order to do so, to run through the pandas dataframe created using a loop and store its values on each google sheet's cell at a time. It's also important to set 120 seconds of time sleep in between this transfer and the next one due to the maximum amount of requests per minute allowed by google. 

```python
Linien_Überspringen = pd.DataFrame(data = Welt_Blatt.get_all_records()).shape[0] + 1
for i in range(len(WeltStat.iloc[0].tolist())):
    Welt_Blatt.update_cell(Linien_Überspringen+1,i+1,int(WeltStat.iloc[0].tolist()[i].replace(",","")))
time.sleep(120)
pd.DataFrame(data = Welt_Blatt.get_all_records())
```

Second data to be stored will be data included corresponding to the top 20 most populated countries in the world. Besides running through the pandas dataframe, it will also be necessary to determine a period for the program to stop for 120 seconds every 3 loops  due to the maximum amount of requests per minute that is allowed by google. After the last loop, there still is another 120 seconds to wait to transfer the last piece of data for the same reason.

```python
Linien_Überspringen = pd.DataFrame(data = Länder_Blatt.get_all_records()).shape[0] + 1
for i in range(len(Völker_Top20Länder.iloc[0].tolist())):
    if (i+1) % 3 == 0:
        time.sleep(120)
    for j in range(20):
        Länder_Blatt.update_cell(Linien_Überspringen+1+j,i+1,Völker_Top20Länder.iloc[j].tolist()[i].replace(",",""))
    
pd.DataFrame(data = Länder_Blatt.get_all_records())
time.sleep(120)
```
